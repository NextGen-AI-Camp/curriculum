## Loop

### While Loop


While Loop เป็น loop แบบหนึ่งที่มีการใช้งานมาก ลักษณะ คือ จะตรวจสอบเงื่อนไข หากเป็น false ก็จะไม่ทำใน code block แต่หากเป็นจริงก็จะทำใน code block ต่อไป ตามรูป
รูปแบบ Syntax มีดังนี้

จากรูปจะเห็นว่า loop เริ่มต้นจาก Test Condition ซึ่งก็คล้ายกับ If statement คือ ตรวจสอบว่า expression เป็นจริงหรือไม่ ถ้าไม่เป็นจริงก็จะข้ามการทำงานใน loop ไป แต่หากเป็นจริง ก็จะทำการทำงานใน loop ดังนั้นเราอาจมองได้ว่า loop ก็คือ If statement ที่มีการวนกลับมาเช็คอีกครั้งก็ได้

ตัวอย่างโปรแกรมนี้จะเป็นการรับข้อมูล และ หาผลรวมของตัวเลขที่รับเข้ามา โดยจะวนทำงานไปจำนวน 5 ครั้ง ตามค่า n ที่กำหนดไว้ (เนื่องจากใน loop จะมีการลดค่า n ลง 1 ทุกครั้ง)

ข้อควรระวังของ loop แบบนี้ คือ เงื่อนไขในการออกจาก loop ต้องมีกรณีที่เกิดขึ้นได้จริง เช่น หากลืมเขียนบรรทัด n = n - 1 ก็จะทำให้เกิดการวนลูปแบบไม่รู้จบได้

````python
### หาผลรวมของข้อมูลที่ป้อนเข้ามา
n = 5
sum = 0
while n > 0:
    data = int(input("Enter your number ="))
    sum = sum + data
    n = n - 1    # update counter

# print the sum
print("The sum is", sum)
````

ตัวอย่าง โปรแกรมนี้รับตัวเลขเพียงตัวเดียว แต่หาผลบวกตั้งแต่ 1 จนถึงค่านั้น

````python
### จงเขียนโปรแกรมแสดงผลตั้งแต่ 1 ถึงตัวเลขที่รับเข้ามาโดยใช้ while
n = int(input("Enter your number: "))
count = 0
sum = 0
while (count <= n):
    print(count)
    count = count + 1
    sum = sum + count
print("Sum = ", sum)
````

การใช้ while loop มักใช้กับการถามเพื่อให้ทำซ้ำอีกรอบ

````python
prompt = "\nPlease enter the name of a city you have visited:"
prompt += "\n(Enter 'quit' when you are finished.) "

city = ''
while city != 'quit':
    city = input(prompt)
    print(f"I'd love to go to {city.title()}!")
````

ในกรณีที่มีเงื่อนไขอื่นที่ต้องการให้ออกจาก loop นอกเหนือจาก test condition สามารถใช้คำสั่ง break เพื่อออกจาก loop ได้เช่นกัน

````python
i = 0
str = "Computer Engineering"

while i < len(str):
    if str[i] == 'e' or str[i] == 's':
        i =  i + 1
        break
    print("current letter = ",str[i])
    i = i + 1
````

ในบางครั้งเราต้องการข้อยกเว้น ไม่ทำงานใน Block Code สำหรับบางเงื่อนไข เราสามารถใช้ continue เพื่อให้ข้ามกลับไปที่ต้น Loop ใหม่


````python
i = 0
str = "Computers Engineering"

while i < len(str):
    if str[i] == 'e' or str[i] == 's':
        i =  i + 1
        continue
    print("current letter = ",str[i])
    i = i + 1
````

ในการทำงานบางอย่างที่ต้องการทำงานซ้ำๆ แทนที่จะเขียนโปรแกรมซ้ำ เราสามารถสร้างเป็น Function ได้

จากรูปอธิบายได้ดังนี้
  - def ย่อมาจาก define แปลว่ากำหนดฟังก์ชัน
  - name เป็นชื่อของฟังก์ชัน
  - argument เป็น input ของฟังก์ชัน (อาจไม่มีก็ได้)
  - statement เป็นสิ่งที่ฟังก์ชันทำ

````python
def greet():
    print('Hello, World!')

#call the function
greet()
greet()

print("Outside function")
````

หลังจากที่กำหนดฟังก์ชัน แล้วเราสามารถเรียกใช้งานได้ โดยลำดับการทำงานของการทำงานของฟังก์ชัน จะเป็นไปตามรูป

ในบรรทัดตั้งแต่ def greet() เป็นต้นไป จนถึงจบฟังก์ชัน จะถือว่าเป็นการนิยาม (define) ฟังก์ชันเท่านั้น จะยังไม่มีการทำงานอะไร แต่เมื่อถึง greet() จึงเป็นการเรียกฟังก์ชันมาทำงาน (1) การทำงานของโปรแกรมจะกระโดดไปที่ (2) และเริ่มทำงานตามฟังก์ชัน จนกระทั่งทำงานตามฟังก์ชันครบหมดแล้วจึงกระโดดกลับมา ทำงานต่อ (3)

กรณีที่มีการส่งค่า argument หรือ parameter จะมีการส่งค่าไปตามลำดับตามภาพ

````python
def is_square(n):
  for i in range(n):
    if i*i == n:
      return True
  return False

print(is_square(25))
print(is_square(5))
````

การส่งค่ากลับ สามารถส่งค่ากลับมากกว่า 1 ค่าก็ได้

ปล. ในการทำงานจริง Python จะส่งกลับข้อมูลเดียว โดยเอาข้อมูล vat และ price มา pack รวมกันเป็น (vat, price)

````python
def price_with_vat(amount):
  vat = amount * 7 / 107
  price = amount - vat
  return vat, price

vat, price = price_with_vat(107)
print(vat, price)
````

ในกรณีที่มี parameter หลายตัว มีความเป็นไปได้ว่าอาจใส่สลับกันมา เพื่อป้องกันไม่ให้ใส่สลับกัน อาจใช้วิธีระบุชื่อตัวแปรตอนเรียกก็ได้ (named argument)

เราสามารถเริ่มใช้ named argument ตอนไหนก็ได้ แต่ argument หลังจากนั้นต้องเป็น named argument ด้วยทั้งหมด (ก่อนหน้านั้น ถือว่าเรียงตามลำดับ)


````python
def price_with_vat(amount, vat_rate):
  vat = amount * vat_rate / (100+vat_rate)
  price = amount - vat
  return vat, price

vat, price = price_with_vat(vat_rate=7, amount=107)
print(vat, price)
````

## Class

Class เป็นแนวคิดเรื่อง Object Oriented Programming ซึ่งมีแนวคิดว่า data และ code ส่วนที่เกี่ยวข้องกัน ควรจับมัดไว้ด้วยกัน

เช่น สมมติว่ามีข้อมูลนักเรียน ประกอบด้วย รหัสนักเรียนและชื่อนักเรียน และ นักเรียนมีการลงทะเบียนในรายวิชา สามารถเขียนเป็น Class ได้ตามโปรแกรมด้านล่าง

- `def __init__` เป็นฟังก์ชันที่ทำงานอัตโนมัติ เมื่อมีการสร้าง object จากรูปจะเห็นได้จาก class dog เมื่อมีการสร้าง object ของสุนัขแต่ละตัว จะต้องบอกว่าเป็น พันธ์อะไร และ อายุเท่าไร ประเด็นคือจะกำหนด พันธ์ และ อายุ ของสุนัขเมื่อไร คำตอบก็คือ กำหนดเมื่อมีการสร้าง object ดังนั้นสรุปได้ว่าการสร้าง object จะต้องกำหนด พันธ์ และ อายุ ประเด็นต่อมา คือ ไปกำหนดที่ไหน คำตอบ คือ ใน `def __init__`

<img src="https://cdn-images-1.medium.com/max/1055/1*NxXw72-CDGp7b7pz2kga5g.png" width="600">









````python
class Student:
    def __init__(self, student_id, student_name):
        self.student_id = student_id
        self.student_name = student_name
        self.subjects = []

    def enroll_subject(self, subject):
        self.subjects.append(subject)

somchai = Student('66010123', 'somchai')
somchai.enroll_subject('Math')
somchai.enroll_subject('Physics')

print(somchai.student_id)
print(somchai.student_name)
print(somchai.subjects)
````

- จากโปรแกรม จะเห็นว่ามีการกำหนด  `def __init__` ซึ่งรับพารามิเตอร์ 2 ตัว คือ student_id และ student_name จากนั้นจึงทำการสร้าง object ของ class อาจกล่าวได้ว่าการสร้าง object ของ class คือ การนำแม่แบบมาสร้างเป็นของจริง ให้นึกถึงพิมพ์ขนมก็ได้ การสร้าง object คือ การนำเอาขนมมาใส่ในพิมพ์ขนมแล้วออกมาเป็นขนม หรือ การสร้างบ้านตามแบบที่กำหนด ซึ่งจะออกมาเป็นบ้านจริงๆ ดังนั้นในบรรทัด

somchai = Student('66010123', 'somchai')

- ก็จะเกิด object ของนักเรียนขึ้นมาจริงๆ

- สำหรับ list self.subjects = [] ใช้ในการเก็บว่านักเรียนคนนั้นเรียนวิชาอะไรบ้าง

- จะเห็นได้ว่า class หรือ object ก็เป็นการรวมเอา data และ code มารวมไว้ด้วยกัน กรณีนี้ data คือ student_id และ student_name ส่วน code ก็คือ ฟังก์ชัน enroll_subject นั่นเอง

- อาจจะสังเกตเห็นว่าตัวแปรของ class จะต่างจากตัวแปรทั่วไป เพราะมีคำว่า self. นำหน้าเสมอ เพื่อให้ทราบว่านี่เป็นตัวแปรของ class หรือ object ไม่ใช่ตัวแปรทั่วไป



รายละเอียดของ OOP หรือ Class ยังมีอีกมาก หากสนใจสามารถศึกษาเพิ่มเติมได้ ในบทนี้จะกล่าวถึงเฉพาะที่จำเป็นต่อการเขียนโปรแกรมในหลักสูตรนี้เท่านั้น
