# Type of Inputs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Using numpy arrays
x_np = np.linspace(0, 10, 100)
y_np = np.sin(x_np)

plt.scatter(x_np, y_np)
plt.title('Plot using numpy arrays')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Using pandas DataFrame
df = pd.DataFrame({'X': x_np, 'Y': y_np})

plt.scatter('X', 'Y', data=df)
plt.title('Plot using pandas DataFrame')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Stack Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
values1 = [10, 20, 30, 40]
values2 = [20, 30, 40, 10]
values3 = [40, 30, 20, 10]

# Plot

plt.stackplot(categories, values1, values2, values3, labels=['Group 1', 'Group 2', 'Group 3'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stack Plot Example')
plt.legend(loc='upper left')

plt.show()
```

# Figure with more than 1 Axes

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot on the first figure
plt.figure()
plt.plot([1, 2, 3], [4, 5, 6])

# Plot on the second figure
plt.figure()
plt.plot([1, 2, 3], [6, 5, 4])

# Create a figure with 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot on the first subplot (top-left)
axs[0, 0].plot([1, 2, 3], [4, 5, 6])

# Plot on the second subplot (top-right)
axs[0, 1].plot([1, 2, 3], [6, 5, 4])

# Plot on the third subplot (bottom-left)
axs[1, 0].plot([1, 2, 3], [2, 3, 1])

# Plot on the fourth subplot (bottom-right)
axs[1, 1].plot([1, 2, 3], [3, 2, 1])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data with dates and strings
dates = pd.date_range('2022-01-01', periods=10)
values = np.random.rand(10)

# Plot
plt.plot(dates, values, marker='o', linestyle='-', color='b')

# Format x-axis with dates and rotate tick labels
plt.xticks(rotation=45)

# Set x-axis label
plt.xlabel('Date')

# Set y-axis label
plt.ylabel('Value')

# Set plot title
plt.title('Value over Time')

# Show plot
plt.show()
```

```python
# Mock up some simple data for a confusion matrix
confusion_matrix = np.array([[20, 5, 0],
                              [3, 25, 2],
                              [1, 4, 15]])

# Define class labels
class_names = ['Class 1', 'Class 2', 'Class 3']

# Plot the confusion matrix
plt.imshow(confusion_matrix, cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()

# Add tick marks and labels
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45)
plt.yticks(tick_marks, class_names)

# Add numerical annotations
for i in range(len(class_names)):
    for j in range(len(class_names)):
        plt.text(j, i, str(confusion_matrix[i, j]), horizontalalignment='center'
        , verticalalignment='center')

plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.tight_layout()

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Mock up some simple data for a confusion matrix
confusion_matrix = np.array([[20, 5, 0],
                              [3, 25, 2],
                              [1, 4, 15]])

# Define class labels
class_names = ['Class 1', 'Class 2', 'Class 3']

# Plot confusion matrix using Seaborn
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)

plt.title('Confusion Matrix')
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
```
