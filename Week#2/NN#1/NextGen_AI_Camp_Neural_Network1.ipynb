## ความเป็นมาของ Perceptron



ในปี 1943 มีนักประสาทวิทยา(Neurophysiologist) ชื่อว่า Warren McCulloch และนักตรรกศาสตร์ ชื่อว่า Walter Pitts ได้พัฒนาแบบจำลองของเซลประสาทเทียม และเขียนบทความวิจัยเรื่อง "A Logical Calculus of the Ideas Immanent in Nervous Activity"

โดยทั้งสองนำเสนอว่า เมื่อเซลล์ประสาทได้รับการกระตุ้นผ่านทาง เดนไดรท์ (Dentrite) มันจะแสดงพฤติกรรมในลักษณะ “all-หรือ-none” กล่าวคือ เมื่อได้รับการกระตุ้นที่แรงมากพอ มันจะส่งสัญญาณต่อไปทาง แอ็กซอน (Axon) แต่หากสัญญาณไม่แรงมากพอ มันจะไม่ส่งสัญญาณออกไปเลย ซึ่งมีลักษณะคล้ายกับ 0 และ 1 ในตรรกศาสตร์

จากสมมติฐานที่กล่าวมาข้างต้น จึงอนุมานต่อไปว่าเหตุการณ์ทางประสาทและความสัมพันธ์ของเหตุการณ์ทางประสาททั้งหมด ควรจะอธิบายด้วยระบบตรรกศาสตร์ได้ด้วย ทั้งสองได้เสนอแบบจำลองของเซลล์ประสาท ซึ่งสามารถนำแบบจำลองนี้มาสร้างเป็นโครงข่ายประสาทเทียมได้

ภาพด้านล่างแสดงการเปรียบเทียบระหว่างเซลล์ประสาทจริง กับแบบจำลองเซลล์ประสาทเทียมที่นำเสนอ ในขณะที่เซลล์ประสาทจริง รับข้อมูลจากเซลล์ประสาทอื่นผ่านทางเดนไดรท์ แต่เซลล์ประสาทเทียมจะรับข้อมูลผ่านทางอินพุต (ซึ่งอาจจะมาจากเซลล์ประสาทเทียมอื่นๆ ก็ได้) และในขณะที่เซลล์ประสาทจริง ส่งข้อมูลออกทางแอ็กซอน แต่เซลล์ประสาทเทียมจะส่งข้อมูลออกทางเอาท์พุต

## การทำงานของ Perceptron

ปัจจุบัน Perceptron ได้ผ่านการพัฒนามาอย่างมากมาย จนกลายเป็น Neural Network กลายเป็น Deep Learning และอื่นๆ ที่ก้าวหน้ามากๆ ตามที่เห็นกันในปัจจุบัน แต่ถึงอย่างไรส่วนประกอบพื้นฐานของ Deep Learning หรือ Nueral Network ก็ยังคงเป็น Perceptron อยู่ดีนั่นเอง

ดังนั้นหากจะทำความเข้าใจกับ Neural Network หรือ Deep Learning อย่างถ่องแท้ คงหลีกเลี่ยงไม่ได้ที่จะทำความเข้าใจกับการทำงานของ Perceptron เสียก่อน

การทำงานของ Perceptron จะอธิบายผ่านการทำงานจริง สมมติว่าเรามีข้อมูลคู่ลำดับ x, y ดังต่อไปนี้ และ มีโจทย์ว่าให้หาความสัมพันธ์ระหว่าง x กับ y (hint เล็กน้อยว่าความสัมพันธ์ระหว่าง x กับ y เป็นความสัมพันธ์แบบเส้นตรง) เราจะมีวิธีในการหาความสัมพันธ์นี้อย่างไร

ลองมาดูการใช้ perceptron ในการแก้ปัญหานี้ ตัวอย่างโปรแกรมต่อไปนี้ เขียนด้วยภาษา Python โดยใช้ Library ชื่อ Pytorch ซึ่งเป็น Library สำหรับทำ Deep Learning ซึ่งเป็นที่นิยมมากที่สุดในเวลานี้ในการเขียน
  

````python
import torch
import torch.nn as nn
import torch.optim as optim
from numpy import *
import matplotlib.pyplot as plt

# Create sample data using PyTorch
x = torch.arange(0, 10, 0.1).unsqueeze(1)
noise = 2 * torch.randn(x.size())
y = 5 * x + 10 + noise

# Define a simple linear model
class LinearModel(nn.Module):
    def __init__(self):
        super(LinearModel, self).__init__()
        self.linear = nn.Linear(1, 1)

    def forward(self, x):
        return self.linear(x)

model = LinearModel()

# Define a loss function and an optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Train the model
num_epochs = 500
losses = []

for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()

    outputs = model(x)
    loss = criterion(outputs, y)

    loss.backward()
    optimizer.step()

    losses.append(loss.item())  # Track the loss

    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# After training, print the learned parameters
[w, b] = model.parameters()
print(f'Learned parameters: w = {w.item():.4f}, b = {b.item():.4f}')
print

# Plot the results
predicted = model(x).detach().numpy()
x_np = x.numpy()  # Convert to numpy for plotting
y_np = y.numpy()  # Convert to numpy for plotting

# Plot the original data and the fitted line
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_np, y_np, 'ro', label='Original data')
plt.plot(x_np, predicted, label='Fitted line')
plt.legend()
plt.title('Fitted Line')

````

### Loss Function
มาถึงตรงนี้ก็ขออธิบายเพิ่มเติมว่า สิ่งที่เป็นเป้าหมายของเราคือ เส้นแนวโน้มของข้อมูล ซึ่งก็คือ y=5x+10 นั่นเอง แต่พึงระลึกว่าขณะนี้สิ่งที่เรามีคือ จุดของข้อมูล x,y เท่านั้น ถ้าดูด้วยสายดา มนุษย์ก็อาจจะพอประมาณได้ว่าสมการเส้นตรงของเส้นแนวโน้มจะเป็นเท่าไร แต่คอมพิวเตอร์ยังทำไม่ได้ เพราะคอมพิวเตอร์ไม่มีความสามารถในการพิจารณาแนวโน้มข้อมูลเหมือนมนุษย์ ดังนั้นเราต้องเพิ่มความฉลาดให้มัน

ขั้นตอนต่อไปที่เราจะทำ คือ เราจะต้องหาค่า Loss ซึ่งก็คือความแตกต่าง ระหว่างค่าของเป้าหมายกับข้อมูลที่มี จากรูปด้านล่าง ค่าของเป้าหมาย คือ ค่าบนเส้นตรง ซึ่งเป็นสิ่งที่เราต้องการหา แต่จุดสีฟ้า คือ ข้อมูลจริงที่เรามี ดังนั้นเราจะต้องหาเส้นตรงมา 1 เส้น ที่ใกล้เคียงกับข้อมูลที่เรามีมากที่สุด

**เส้นตรงที่ใกล้เคียงกับข้อมูลที่เรามี คือ อะไร **

คือเส้นตรงข้อมูลที่เมื่อรวมเอาค่า Error หรือ ความแตกต่างระหว่างแต่ละจุดกับค่าที่ได้จากสมการเส้นตรง เมื่อผลรวมของค่า Error ของทุกจุดรวมกันแล้ว มีค่าน้อยที่สุด เราก็จะได้เส้นตรงที่ใกล้เคียงกับข้อมูลที่มีมากที่สุด (Best-fit line)

ในการหาค่า Error นั้น ทำได้ง่ายๆ โดยการนำค่า y(จากข้อมูล) - y(เป้าหมายบนเส้นตรง) ซึ่งมักจะเขียน y(เป้าหมาย) เป็น $y'$ หรือ $\hat{y}$ ซึ่งก็จะมีประเด็นอยู่ว่า ผลการลบจะมีทั้งบวกและลบ ซึ่งอาจจะหักล้างกันเอง ทำให้เกิดความผิดพลาดได้ ดังนั้นการแก้ไข คือ จะต้องทำให้ข้อมูลเป็นบวกอย่างเดียว โดยจะเก็บเฉพาะค่าความต่าง

จากความรู้ทางคณิตศาสตร์ การจะทำให้ข้อมูลเป็นบวก สามารถทำได้ 2 วิธี คือ ใช้ absolute (ABS) หรือ ใช้การยกกำลัง 2 (square) ก็ได้ ซึ่งก็ทำให้เกิดวิธีการ 2 วิธี เรียกว่า

1) Sum of Absolute Error หรือ SAE ซึ่งเขียนเป็นสูตรได้ดังนี้

$$ \text{SAE} = \sum_{i=1}^n | y_i - \hat{y}_i | $$

2) Sum of Square Error หรือ SSE ซึ่งเขียนเป็นสูตรได้ดังนี้

$$ \text{SSE} = \sum_{i=1}^n ( y_i - \hat{y}_i )^2 $$

เอาละเมื่อได้สมการมาแล้ว คราวนี้เราจะนำสมการมาเขียนเป็นโปรแกรม ซึ่งจะเขียนเป็นฟังก์ชัน 2 ฟังก์ชันสำหรับ Error แต่ละแบบ ดังนี้



````python

def compute_SAE_for_line_given_points(b, a, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += abs(y - (a * x + b))
    return totalError / float(len(points))

def compute_SSE_for_line_given_points(b, a, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (a * x + b)) ** 2
    return totalError / float(len(points))
````

จากโปรแกรมข้างต้น จะเห็นว่าฟังก์ชันทั้ง 2 ก็เพียงแต่นำค่า y และ y' มาลบกัน จากนั้นก็ทำ Absolute หรือ ยกกำลัง 2 ขึ้นอยู่กับวิธีการ แล้วก็นำข้อมูล Error ของแต่ละจุดมารวมกัน

คำถามต่อมา คือ เมื่อรวม Error ทั้งหมดแล้ว จะเอาไปทำอะไรต่อ ซึ่งเราต้องเข้าใจว่า ขณะนี้เรากำลังหาค่าตัวแปร a และ b ในสมการ

$$ y = ax + b $$

โดยหวังว่าเมื่อหาสมการได้แล้ว มันจะใกล้เคียงสมการเริ่มต้น คือ y = 5x+10

ดังนั้นวิธีการเป็นดังนี้

1.   สุ่มค่า a และ b
2.   นำค่า x ที่ใช้ในสมการ y = ax+b มาหาค่า y' ของทุกจุดของ x
3.   นำค่า y' ของทุกจุดของ x มาเทียบกับค่า y ของข้อมูลในแต่ละจุด และ หา SAE หรือ SSE
4.   นำค่าที่ได้จาก SAE หรือ SSE มาปรับค่า a และ b ที่สุ่มเอาไว้ตั้งแต่เริ่มต้น และ กลับไปทำข้อ 2 อีกครั้ง

จากนั้นก็ทำตามข้อ 2-4 ไปเรื่อยๆ จนกว่าค่า Loss หรือ SAE หรือ SSE จะลดลงจนถึงจุดที่ยอมรับได้

เราจะเริ่มจากข้อ 1 คือ สุ่มค่า a และ b ซึ่งโดยอำนาจเผด็จการ ผมสุ่มเอาเอง โดยให้ a เป็น 20 และ b เป็น 30

สำหรับข้อ 2 และ 3 ได้เขียนโปรแกรมไว้แล้ว ในฟังก์ชัน compute_SAE_for_line_given_points หรือ  compute_SSE_for_line_given_points

คราวนี้ก็มาถึงข้อ 4 จะปรับค่ายังไงดี ผมจะเริ่มด้วยวิธีการง่ายๆ และอาจจะไม่ค่อยฉลาด คือ จะลดค่า a และ b ไปเรื่อยๆ ทีละ 0.1 จนกว่าจะได้ค่า Loss ต่ำที่สุด สามารถเขียนเป็นโปรแกรมดังนี้


````python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate some points
def generate_points(num_points=100):
    np.random.seed(0)
    x = np.arange(0, 10, 0.1)
    x = np.expand_dims(x, 1)
    noise = random.randn(*x.shape)
    y = 5*x+10 + noise
    points = concatenate((x,y), axis=1)

    return points

# Compute SAE function
def compute_SAE_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += abs(y - (m * x + b))
    return totalError / float(len(points))

# Generate some example points
points = generate_points()

# Parameters
initial_a = 20  # initial slope
initial_b = 30  # initial y-intercept
step_size = 0.1
num_iterations = 50  # number of iterations

# Find minimum SAE
best_a = initial_a
best_b = initial_b
min_sae = compute_SAE_for_line_given_points(best_b, best_a, points)

for i in range(num_iterations):
    # Decrement a and b
    current_a = best_a - (i * step_size)
    current_b = best_b - (i * step_size)

    # Compute SAE for the current parameters
    current_sae = compute_SAE_for_line_given_points(current_b, current_a, points)

    # Check if the current SAE is less than the minimum SAE found so far
    if current_sae < min_sae:
        min_sae = current_sae
        best_a = current_a
        best_b = current_b

print(f"Best slope (a): {best_a}")
print(f"Best y-intercept (b): {best_b}")
print(f"Minimum Sum of Absolute Errors (SAE): {min_sae}")

plt.plot(x,y,'bo')
plt.plot(x, best_a*x+best_b,'r')
plt.show()

````

### Gradient

คราวนี้จะทำอย่างไรดี มีผู้ที่นำหลักการทางคณิตศาสตร์มาใช้แก้ปัญหาเรื่องนี้  โดยนำหลักการของ Calculus ในเรื่อง สมการเชิงอนุพันธ์ (differential equation) มาใช้
ก่อนอื่นขอแนะนำให้รู้จักกับคำว่า Least Square Method เสียก่อน โดย Least Square Method คือ วิธีการที่ใช้ในการหาค่าต่ำสุดของฟังก์ชันใดๆ โดยวิธีการนี้กล่าวถึงคุณสมบัติของจุดต่ำสุดว่า

**"ที่จุดต่ำสุดของฟังก์ชันใดๆ gradient ของฟังก์ชันที่จุดนั้นจะมีค่าเป็น 0"**

แล้ว gradient คืออะไร ถ้าดูจากรูปด้านล่าง จะเห็นว่านิยามของ gradient คือ ความเปลี่ยนแปลงของ y เทียบกับความเปลี่ยนแปลงของ x นั่นก็คือ dy/dx นั่นเอง

ดังนั้นหากเราต้องการหาค่าต่ำสุดของ Loss Function ที่เลือก (บางครั้งก็เรียกว่า Cost Function) ก็แปลว่าเราจะต้องหา gradient (diff) ของฟังก์ชัน คราวนี้ระหว่าง 2 ฟังก์ชัน คือ MAE และ MSE เราจะเลือกใช้ฟังก์ชันใดดี เนื่องจากการหา gradient คือการ diff นั้น จะเหมาะสมสำหรับฟังก์ชันที่มีความต่อเนื่องมากกว่า ซึ่ง MAE เป็นฟังก์ชันที่มีความไม่ต่อเนื่อง เนื่องจากใช้ Absolute ดังนั้นโดยทั่วไปจึงมักเลือก Sum of Square Error หรือ MSE มาใช้เป็น Loss Function หรือ Cost Function มากกว่า


จากสมการ Sum of Squared Errors (SSE) ดังนี้ (บางครั้งอาจมีการใช้สลับระหว่าง y hat กับ y')


$$ \text{SSE} = \sum_{i=1}^n ( y_i - \hat{y}_i )^2 $$

โดยที่:
- $\ y_i\$  เป็นค่าจริงที่ได้มาจากข้อมูล

- $\ \hat{y} \$ (อ่านว่า y-hat) เป็นข้อมูลเป้าหมายจากสมการเชิงเส้น

- n คือ จำนวนข้อมูล

เมื่อแทนค่า $\ \hat{y}_i \ ด้วย \ ax_i + b \$ จะได้

$$ \text{SSE} = \sum_{i=1}^n ( y_i - (ax_i + b) )^2 $$

เนื่องจากสำหรับข้อมูลแต่ละตัว $\ y_i\$ ถือเป็นค่าคงที่ได้ ดังนั้นสมการข้างต้น ถือว่าเป็นสมการ 2 ตัวแปร และ แทนที่เราจะใช้ diff ทั้งสมการ เราจะทำ partial derivative หรือ partial diff กับ SSE

ดังนั้น partial derivative ของ SSE เมื่อเทียบกับ  $\ a \$ คือ:


$$ \frac{\partial \text{SSE}}{\partial a} = \frac{\partial}{\partial a} \sum_{i=1}^n ( y_i - (ax_i + b) )^2 $$

จะได้เท่ากับ

$$ \frac{\partial \text{SSE}}{\partial a} = \sum_{i=1}^n 2 ( y_i - (ax_i + b) ) \cdot (-x_i) $$

ซึ่งลดรูปได้เป็น:

$$ \frac{\partial \text{SSE}}{\partial a} = -2 \sum_{i=1}^n x_i ( y_i - (ax_i + b) ) $$



และ partial derivative ของ SSE เมื่อเทียบกับ  $\ b \$ คือ:

$$ \frac{\partial \text{SSE}}{\partial b} = \frac{\partial}{\partial b} \sum_{i=1}^n ( y_i - (ax_i + b) )^2 $$

จะได้เท่ากับ:

$$ \frac{\partial \text{SSE}}{\partial b} = \sum_{i=1}^n 2 ( y_i - (ax_i + b) ) \cdot (-1) $$

ซึ่งลดรูปได้เป็น:

$$ \frac{\partial \text{SSE}}{\partial b} = -2 \sum_{i=1}^n ( y_i - (ax_i + b) ) $$








### Gradient Descent

ในตอนที่ผ่านมา เราได้อธิบาย Gradient และความสำคัญของ Gradient อย่างไรก็ตาม ค่าที่คำนวณได้เป็นเพียงการวัด gradient หรือ ความชัน ณ จุดที่เป็นอยู่ในปัจจุบันเท่านั้น แต่สิ่งที่ต้องการคือ ค่าของคำตอบ (ในกรณีนี้ คือ ค่า a และ b) ที่ทำให้ความชัน หรือ gradient มีค่าต่ำที่สุด ซึ่งหากจะใช้วิธีการทางคณิตศาสตร์เพื่อหาแล้ว จะมีความซับซ้อนค่อนข้างมาก เช่น หากต้องการหาค่า a ที่จุดต่ำสุด เมื่อแก้สมการแล้ว จะได้คำตอบ ดังนี้


$$ a = \frac{n \sum (x_i y_i) - \sum x_i \sum y_i}{n \sum x_i^2 - (\sum x_i)^2} $$


และ หากต้องการหาค่า b ที่จุดต่ำสุด เมื่อแก้สมการแล้ว จะได้คำตอบ ดังนี้


$$ b = \frac{\sum y_i - a \sum x_i}{n} $$


ซึ่งต้องไม่ลืมว่า นี่เป็นเพียงสมการเชิงเส้นอย่างง่ายที่มีตัวแปรเพียง 2 ตัว เท่านั้น แต่ใน Nueral Network ที่ใช้งานจริง อาจมีตัวแปรหลายล้านตัวแปรก็ได้ ดังนั้นจะเห็นได้ว่าวิธีการทางคณิตศาสตร์ไม่เหมาะสมที่จะนำมาใช้ในกรณีนี้ เนื่องจากมีความซับซ้อนมากเกินไป

### Gradient Descent in one dimension (1D)

เพื่อให้เห็นภาพของ gradient descent ได้ชัดเจนขึ้น จะขอยกตัวอย่าง โดยสมมติว่ามี line ข้อมูลตามสูตร y = (x-60)^2+50 ซึ่งเมื่อหาอนุพันธ์จะได้เท่ากับ 2*(x-60) จากนั้นจะลอง plot จุดของแต่ละ step ของ gradient descent โดยทำทั้งหมด 50 ครั้ง จะเห็นได้ว่าในขณะที่ความชันมาก step ของ gradient descent จะมาก แต่เมื่อความชันลดลง step ของ gradient descent ก็จะลดลงด้วย

````python
import numpy as np
import matplotlib.pyplot as plt

# y = (x-60)^2+50

def line_equation(x):
    return (x - 60)**2 + 50

def line_derivative(x):
    return 2 * (x - 60)

def gradient_descent_1d(x_init, learning_rate, epochs):
    x = x_init
    x_values = [x]  # To store x values for visualization
    for _ in range(epochs):
        gradient = line_derivative(x)
        x = x - learning_rate * gradient
        x_values.append(x)  # Append x for visualization
    return x_values

initial_x = 100
learning_rate = 0.1
num_epochs = 50
result_1d = gradient_descent_1d(initial_x, learning_rate, num_epochs)

iterations = np.arange(num_epochs + 1)
plt.plot(iterations, result_1d, marker='o')
plt.xlabel('Iterations')
plt.ylabel('x')
plt.title('Gradient Descent: x Optimization')
plt.grid(True)
plt.show()
````

ลอง list ข้อมูลให้เห็นความห่างของแต่ละ step จะเห็นว่าเนื่องจากความห่างของแต่ละ step ได้มาจากการหาอนุพันธ์ ณ จุดนั้น จึงทำให้ความห่างของแต่ละ step เปลี่ยนแปลงตามความชัน

** ให้ลองเปลี่ยน learning_rate เป็นค่าอื่นๆ ** แล้วดูผลการทำงาน

### Gradient Descent in two dimension (2D)

คราวนี้มาดู gradient descent แบบ 2 มิติ หรือ 2 ตัวแปรดูบ้าง ซึ่งหลักการจะเหมือนกับแบบ 1 มิติ เพราะในการหาอนุพันธ์ เราจะทำทีละมิติ เพียงแต่นำข้อมูลมา plot ให้เห็นเป็น 2 มิติ

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# y = (x1 - 60)**2 + (x2 - 30)**2 + 50
def line_equation_2d(x1, x2):
    return (x1 - 60)**2 + (x2 - 30)**2 + 50

def line_derivative_axis_x1(x1):
    return 2 * (x1 - 60)

def line_derivative_axis_x2(x2):
    return 2 * (x2 - 30)

# 2D gradient descent
def gradient_descent_2d(x1_init, x2_init, learning_rate, epochs):
    x1 = x1_init
    x2 = x2_init

    x1_values = [x1] # lines for visualization
    x2_values = [x2]# lines for visualization
    y_values = [line_equation_2d(x1, x2)] # lines for visualization

    for _ in range(epochs):
        grad_x1 = line_derivative_axis_x1(x1)
        grad_x2 = line_derivative_axis_x2(x2)
        x1 = x1 - learning_rate * grad_x1
        x2 = x2 - learning_rate * grad_x2
        x1_values.append(x1)
        x2_values.append(x2)
        y_values.append(line_equation_2d(x1, x2))

    return x1_values, x2_values, y_values

# you can select any other initial values.
initial_x1 = 70
initial_x2 = 40
learning_rate = 0.1
num_epochs = 50
x1_vals, x2_vals, y_vals = gradient_descent_2d(initial_x1, initial_x2, learning_rate, num_epochs)

# Plotting the 3D graph
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

speed_vals = np.array(x1_vals)
load_vals = np.array(x2_vals)
efficiency_vals = np.array(y_vals)

ax.plot(speed_vals, load_vals, efficiency_vals, marker='o')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Function Value ')
ax.set_title('Gradient Descent in 2D: x1 vs x2 for y results')

plt.show()
````

จากภาพจะเห็นได้ว่า การหา gradient descent นั้นไม่ว่าจะมีกี่มิติ จะพิจารณาทีละมิติเสมอ

## แปลง Gradient Descent เป็นโปรแกรม

จากที่กล่าวมาทั้งหมด เราสามารถนำมาเขียนเป็นโปรแกรมภาษา Python เพื่อใช้ในการปรับพารามิเตอร์ 1 ครั้ง โดยตั้งชื่อว่า step_gradient การทำงานมีดังนี้
- กำหนด gradient เริ่มต้นของ a และ b ให้เป็น 0 เพื่อลบค่า gradient ของรอบก่อนหน้านี้ออกไป
- หาจำนวนข้อมูล (N)
- วนลูปเพื่อหาค่า Gradient ของแต่ละข้อมูลจนครบทุกข้อมูล โดยใช้สูตรของ gradient ที่ได้จากการคำนวณก่อนหน้านี้ โดย a = $-2/N * x * ( y_i - (ax_i + b) ) $ และ b = $-2/N * ( y_i - (ax_i + b) ) $
- จะได้ต่า Gradient สะสมของรอบนั้น
- จากนั้นนำ Gradient ที่ได้ของ a และ b มาปรับค่า a และ b โดยลดขนาดจากการคูณกับ Learning Rate

อาจมีคำถามว่า Learning Rate ควรเป็นเท่าใด เรื่องนี้ไม่มีคำตอบที่ชัดเจน แต่อาจคิดง่ายๆ ว่าหากตั้งค่า Learning Rate เป็น 0.01 ก็เท่ากับปรับ Gradient รอบละประมาณ 1% หากหากตั้งค่า Learning Rate เป็น 0.001 ก็เท่ากับปรับ Gradient รอบละประมาณ 0.1% การใช้ค่า Learning Rate น้อยๆ ไม่มีผลเสียอะไร นอกจากทำให้การทำงานนานขึ้น การหาจุดต่ำสุดก็จะนานไปด้วย แต่หากใช้ค่า Learning Rate มากเกินไปอาจทำให้กระโดดข้ามจุดต่ำสุดตามรูปขวาด้านบน ซึ่งจะทำให้ไม่สามารถหาจุดต่ำสุดได้

````python
def step_gradient(a_current, b_current, points, learning_rate):
    a_gradient = 0
    b_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        a_gradient += -(2/N) * (x * (y - ((a_current * x) + b_current)))
        b_gradient += -(2/N) * ((y - ((a_current * x) + b_current)))
    new_a = a_current - (learning_rate * a_gradient)
    new_b = b_current - (learning_rate * b_gradient)
    return [new_a, new_b]
````

เมื่อได้ฟังก์ชันที่ทำหน้าที่ปรับ gradient ในแต่ละรอบการทำงานแล้ว (จะเรียกรอบการทำงานที่ประมวลผลข้อมูลครบทุกข้อมูลนี้ว่า epoch) ขั้นตอนต่อไปเราจะเขียนโปรแกรมที่ทำหน้าที่เรียกใช้ step_gradient อีกที โดยทำหน้าที่วนลูปเรียกจนครบตามจำนวนรอบ (epoch) ที่กำหนดไว้ โดยจะเก็บค่า Loss ของแต่ละรอบ เพื่อนำมา plot graph เอาไว้ด้วย

## โปรแกรมของทุกส่วน
เมื่อได้โปรแกรมครบทุกส่วนแล้ว คราวนี้ก็นำมาประกอบร่างและดูว่าจะทำงานได้ผลดีหรือไม่

````python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate some points
def generate_points(num_points=100):
    np.random.seed(0)
    x = np.arange(0, 10, 0.1)
    x = np.expand_dims(x, 1)
    noise = np.random.randn(*x.shape)
    y = 5*x+10 + noise
    points = np.concatenate((x,y), axis=1)

    return points

def compute_SSE_for_line_given_points(a, b, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (a * x + b)) ** 2
    return totalError / float(len(points))

def step_gradient(a_current, b_current, points, learning_rate):
    a_gradient = 0
    b_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        a_gradient += -(2/N) * (x * (y - ((a_current * x) + b_current)))
        b_gradient += -(2/N) * ((y - ((a_current * x) + b_current)))
    new_a = a_current - (learning_rate * a_gradient)
    new_b = b_current - (learning_rate * b_gradient)
    return [new_a, new_b]

def gradient_descent_runner(points, starting_a, starting_b, learning_rate,
                            num_iterations):
    a = starting_a
    b = starting_b
    loss = []
    for i in range(num_iterations):
        a, b = step_gradient(a, b, np.array(points), learning_rate)
        loss.append(compute_SSE_for_line_given_points(a, b, points))

    return a, b, loss

# Generate some example points
points = generate_points()

# Parameters
initial_a = 20  # initial slope guess
initial_b = 30  # initial y-intercept guess
learning_rate = 0.0003  # learning rate
num_iterations = 20000 # number of iteration, epoch

print("Starting gradient descent at a = {0}, b = {1}, error = {2}".format(initial_a, initial_b,
compute_SSE_for_line_given_points(initial_a, initial_b, points)))
print("Running...")
a, b , loss = gradient_descent_runner(points, initial_a, initial_b, learning_rate, num_iterations)
print("After {0} iterations a = {1}, b = {2}, error = {3}".format(
    num_iterations, a, b, compute_SSE_for_line_given_points(a, b, points)))
print(len(loss))

# Extract x and y from points for plotting
x = points[:, 0]
y = points[:, 1]

# Plot the points and the fitted line
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(x, y, color='blue')
plt.plot(x, a * x + b, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression with Gradient Descent')

# Plot the loss graph
plt.subplot(1, 2, 2)
plt.plot(range(num_iterations), loss, label='Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss over Epochs')
plt.legend()

plt.tight_layout()
plt.show()


````
