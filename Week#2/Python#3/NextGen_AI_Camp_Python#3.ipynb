### Slicing

Python มีความสามารถในการระบุสมาชิกเป็นช่วง เรียกว่า Slicing ซึ่งมีการใช้งานมากใน Machine Learning โดยเฉพาะ Computer Vision หรือ Image Processing ดังนั้นหากจะเอาดีทางนี้ ควรฝึกฝนการทำ Slicing

ในการ Slicing จะมีตัวเลข 3 ตัว คือ start, stop, step
- start คือ ตำแหน่งเริ่มต้น จะใช้แบบเลขบวก หรือ เลขลบก็ได้
- stop คือ ไม่เกินตำแหน่งนี้ ดังนั้นตำแหน่งสุดท้ายจริงคือ stop -1
- step คือ การกระโดด

````python
list2 = [1, 2, 3, 4, 5, 6, 7];
print("list2[1:5]:", list2[1:5])
print("list2[::2]:", list2[::2])
print("list2[2::2]:", list2[2::2])
print("list2[2:7:2]:", list2[2:7:2])
print("list2[:7]:", list2[:7])        # from begin
print("list2[4:]:", list2[4:])        # to end
````

````python
lst = [40, 50, 20, 30, 90, 80, 70]
print(lst[1:3])
print(lst[2:6])
print(lst[0:8:3])
print(lst[-2])
print(lst[:-2:2])
print(lst[::4])
print(lst[2:-2])
print(lst[::-1])
print(lst[::-2])
print(lst[:-2:-2])
````

### การเพิ่มข้อมูลใน List

สามารถทำได้หลายวิธี

- ใช้เครื่องหมาย + โดยข้อมูลที่จะนำมา + ต้องเป็น list

````python
flowers = ['Rose', 'Lily', 'Tulip', 'Carnation', 'Poppy', 'Sunflower']
flowers += ['Magnolia']
print(flowers)
````

- ใช้ฟังก์ชัน append หรือ extend

````python
flowers = ['Rose', 'Lily', 'Tulip', 'Carnation', 'Poppy', 'Sunflower']
flowers.append(['Iris'])
print(flowers)
flowers.extend(['Violet','Forget me not'])
print(flowers)
flowers.append('Jasmine')
print(flowers)
flowers.extend('Lavender')
print(flowers)
flowers.append('Jasmine')
print(flowers)
````

````python
flowers = ['Rose', 'Lily', 'Tulip', 'Carnation', 'Poppy', 'Sunflower']
flowers += 'Rosemary'
print(flowers)
````

### การแก้ไขข้อมูลใน List

การแก้ไขข้อมูลใน List จะเป็นการกำหนดค่าใหม่ให้กับแต่ละตำแหน่งที่อยู่ใน List โดยตรง

````python
list1 = ['physics', 'chemistry', 'calculus', 'biology'];
print("Old value available at index 2 : ", list1[2])
list1[2] = 'programming';
print("New value available at index 2 : ", list1[2])
print(list1)
````

````python
odd = [2,4,6,8]
odd[1:4] = [3,5,6,7,8]
odd
````

### การแทรกข้อมูล

การแทรกข้อมูลใน List จะใช้คำสั่ง insert โดยระบุตำแหน่งที่ insert

insert(position, value)


````python
motorcycles = []
motorcycles.append('honda')
motorcycles.append('yamaha')
motorcycles.append('suzuki')
print(motorcycles)
motorcycles.insert(0, 'ducati')
print(motorcycles)
````

### การลบข้อมูลใน List

การลบข้อมูลใน List สามารถทำได้หลายวิธี
- ใช้คำสั่ง clear เป็นการลบ list ทั้งหมด
- ใช้คำสั่ง remove เป็นการลบโดยอ้างถึงข้อมูลที่ต้องการลบ
- ใช้คำสั่ง del เป็นการลบโดยอ้างถึงตำแหน่งที่ต้องการลบ
- ใช้คำสั่ง pop หากไม่ระบุตำแหน่ง จะเป็นการลบข้อมูลท้ายสุด แต่ถ้าระบุตำแหน่งจะเป็นการลบในตำแหน่งที่ระบุ

````python
motorcycles = ['honda', 'yamaha', 'suzuki', 'ducati']

print(motorcycles)
del motorcycles[2]
print(motorcycles)
motorcycles.pop()
print(motorcycles)
````

````python
motorcycles = ['honda', 'yamaha', 'suzuki', 'ducati']

print(motorcycles)
del motorcycles[2]
print(motorcycles)
motorcycles.pop(0)
print(motorcycles)
motorcycles.remove('ducati')
print(motorcycles)
motorcycles.clear()
print(motorcycles)
````

### การเปรียบเทียบ List

- list สามารถ เปรียบเทียบได้ด้วย โดยการเปรียบเทียบจะเปรียบเทียบเรียงตามลำดับ
 ดังนั้น [1,2] จะไม่เท่ากับ [2,1]

````python
lst1 = [9, 2, 3, 4, 5]
lst2 = [9, 1, 7, 6, 5]
lst3 = [9, 8, 7, 6, 5]
lst4 = [8, 7]
print("lst1 < lst2 :", lst1 < lst2)
print("lst1 > lst2 :", lst1 > lst2)
print("lst2 >= lst1 :", lst2 >= lst1)
print("lst2 == lst3 :", lst2 == lst3)
````

### การสร้าง List ชนิดลำดับตัวเลข

- กรณีที่ต้องการสร้างข้อมูลตัวเลขที่มีลำดับที่แน่นอน เราสามารถใช้คำสั่ง range(start, stop, step)
- start = เริ่ม, stop = สิ้นสุด (ไม่รวม stop), step = กระโดดครั้งละ



````python
even_numbers = list(range(2, 11, 2))
print(even_numbers)

numbers = list(range(1, 6))
print(numbers)
````

### ฟังก์ชันที่เกี่ยวกับ List อื่นๆ
- x.sort() ทำให้ข้อมูลในลิสต์ x เรียงจากน้อยไปมาก คำสั่งนี้ไม่มี return คืนกลับมา
- sorted(x) คืนลิสต์ที่มีค่าเหมือนกับใน x แต่เรียงลำดับข้อมูลจากน้อยไปมากให้
(x ไม่เปลี่ยนแปลง)
- sum(x) คืนผลรวม (summary) ของตัวเลขที่อยู่ในลิสต์ x
- max(x) คืนค่ามากสุดในลิสต์ x, min(x) คืนค่าน้อยสุดในลิสต์ x
- x.count(e) คืนจำนวนครั้งที่ e ปรากฏในลิสต์ x


### Nested List

การอ้างถึงสมาชิกใน matrix มีดังนี้



### การใช้ Slicing ใน String
