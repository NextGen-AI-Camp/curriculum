## ข้อมูลชนิด Boolean

ประโยค a = “1” == True มีค่าเป็นอะไร

- การกระทำแบบเปรียบเทียบ สามารถเขียนในรูปแบบ “ลูกโซ่” หรือ Chained Comparison ได้

- การเขียนในลักษณะ “ลูกโซ่” แบบนี้ จะทำให้ นิพจน์ สั้นลง

````python
print(5 < 6 > 2)
print(5 < 6 and 6 > 2	)
print(1 < 2 < 3 < 4)
print(1 < 2 and 2 < 3 and 3 < 4)
````

## การกระทำทาง Logic

การกระทำทาง Logic คือ การกระทำระหว่างข้อมูลที่เป็น Boolean ด้วยกัน และให้ผลลัพธ์เป็น Boolean มีทั้งหมด 4 การกระทำ

- Operator XOR จะ True กรณีที่ไม่เหมือนกัน

## If statement

If statement เป็นการทำงานที่พบได้บ่อยในโปรแกรม โดยเป็นการกำหนดให้โปรแกรมทำงานใดทำงานหนึ่งตามเงื่อนไข เช่น
- ถ้าฝนตก ให้กางร่ม
- ถ้าสูงน้อยกว่า 100 cm ให้เข้าฟรี และ ถ้าสูงน้อยกว่า 150 cm เสียครึ่งราคา

กล่าวคือ การทำงานใดที่เป็นลักษณะของเงื่อนไข จะใช้ if statememt

มีรูปแบบการใช้งานดังนี้
- เริ่มด้วย if
- ต่อด้วย condition expression ซึ่งต้องให้ผลเป็น True หรือ False
- ปิดท้าย expression ด้วยเครื่องหมาย :
- Block code ส่วนของการทำงานเมื่อเป็น True (ต้อง Indent)

> <img src="https://www.learnbyexample.org/wp-content/uploads/python/Python-if-Statement-Syntax.png" width="400">

จากรูปจะเห็นได้ว่าการทำงาน คือ จะตรวจสอบเงื่อนไขที่ใส่ใน expression หลัง if ถ้าเงื่อนไขเป็นจริง ก็จะทำใน statement ที่อยู่ใน block code (สีเขียว) แต่ถ้าเงื่อนไขเป็นเท็จ ก็จะข้ามการทำงานสีเขียวไป



````python
# สมมติว่าค่าเข้าสวนสนุก เด็กจะได้ลด 50% ถ้าสูง < 120

TICKET_PRICE = 100
payment = TICKET_PRICE
height = int(input("What is your height in cm. :"))
if height < 120:
    payment *= 0.5
print(f"Your payment is {payment}")
````

## if else statement

- บางครั้งจะมีเงื่อนไขที่ถ้า ใช่ ให้ทำแบบหนึ่ง  แต่ถ้าไม่ใช่ให้ทำอีกแบบหนึ่ง
- จากรูปจะเห็นว่าถ้า Expression เป็น True จะทำ Block code แรก (สีเขียว) แต่ถ้าเป็น False ก็จะทำ Block code ที่ 2 (สีม่วง)
- จะมีคำว่า else: เพิ่มเข้ามา

<br>

> <img src="https://www.learnbyexample.org/wp-content/uploads/python/Python-if-else-Statement-Syntax.png" width="400">


````python
# เลขคู่หรือเลขคี่

num = int(input("Enter your number (integer): "))
if num % 2 == 0:
    print(num, "is an even number")
else:
    print(num, "is an odd number")
````

## elif statement

จะใช้ในกรณีที่มีเงื่อนไขหลายเงื่อนไขที่เกี่ยวข้องกันให้ตรวจสอบ เช่น
- การเก็บค่าผ่านประตูเข้าอุทยานแห่งชาติ ที่มีทั้งเงื่อนไข คนไทย และ ต่างชาติ เด็ก และ ผู้สูงอายุ ที่เก็บไม่เท่ากัน
- การเก็บเงินค่าตั๋วหนังที่ไม่เท่ากันในเก้าอี้แต่ละแบบ

รูปแบบการเขียนจะมีดังนี้
- ในเงื่อนไขแรก จะเขียนเหมือน if else ตามปกติ
- ในเงื่อนไขที่ 2 และต่อ ๆ ไป จะเขียนเป็น elif ซึ่งสามารถใช้ได้หลายครั้ง
- และหากไม่ตรงกับเงื่อนไขใดเลย จะทำงาน ใน Block else สุดท้าย
- ในส่วนของ else สุดท้าย จะมีหรือไม่ก็ได้

<img src="https://www.learnbyexample.org/wp-content/uploads/python/Python-elif-Statement-Syntax.png" width="400">

````python
### ตัวอย่าง : ในสวนสนุกแห่งหนึ่งมีอัตราการเข้าชมต่างกันตามอายุ
### ถ้าต่ำกว่า 4 ขวบ ฟรี
### อายุ 4-18 = 25 บาท
### อายุ 18-65 = 40 บาท
### อายุ > 65 = 20 บาท
### จงเขียนโปรแกรมรับอายุแล้วคำนวณจำนวนเงิน

age = int(input("Enter your age : "))

if age <= 4:
    price = 0
elif age <= 18:
    price = 25
elif age < 65:
    price = 40
elif age >= 65:
    price = 20

print(f"Your admission cost is {price}.")
````

````python
#-------------------------------------------
#* หรือจะเขียนแบบนี้ก็ได้
#-------------------------------------------

age = int(input("Enter your age : "))

if age <= 4:
    price = 0
if 4 < age <= 18:
    price = 25
if 18 < age <= 65:
    price = 40
if age >= 65:
    price = 20

print(f"Your admission cost is {price}.")
````

````python
### ตัวอย่าง : สมมติว่าการโอนเงินไปต่างประเทศมีค่าธรรมเนียม
###          0-5000 ไม่มีค่าธรรมเนียม
###          5000-30,000        2 บาท
###          30,000-100,000     5 บาท
###          >100,000           10 บาท
###          จงหาจำนวนเงินที่ต้องจ่ายในการโอน

amount = int(input("Enter amount of transfer money: "))
fee = 0
if amount<=5000:
    fee=0
elif amount <= 30000:
    fee=2
elif amount <= 100000:
    fee=5
else:
    fee=10

print(f"Transfer fee = {fee}.")
````

## Nest If

คำสั่ง if  บางครั้งอาจมีเงื่อนไขที่ต้องพิจารณามากกว่า 1 เงื่อนไข
เช่น สมมติว่า การเข้าอุทยานแห่งชาติ มีการแบ่งออกเป็น
- คนไทย    ผู้ใหญ่ 50 บาท เด็ก 20 บาท
- ต่างชาติ   ผู้ใหญ่ 100 บาท เด็ก 40 บาท

จะเห็นว่ามี 2 เงื่อนไข คือ คนไทย หรือ ต่างชาติ กับ เด็ก หรือ ผู้ใหญ่ แต่คำตอบต้องเป็นคำตอบเดียวกัน



````python
thai_citizen = input("Thai citizen (Y/N) :")
age = int(input("Enter your age : "))

fee=0
if (thai_citizen=='Y'):
    if (age < 18):
        fee = 20
    else:
        fee = 50
else:
    if (age < 18):
        fee = 40
    else:
        fee = 100

print(f"Your admission cost is {fee}.")
````
